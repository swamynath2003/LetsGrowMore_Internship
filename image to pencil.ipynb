```python
import os

if not os.path.exists(image_path):
    raise FileNotFoundError(f"Image not found at {image_path}")
```

```python
import cv2

def pencil_sketch(image_path):
    # Read the image in RGB format
    image = cv2.imread(image_path)
    # Convert the image to grayscale
    gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
    # Invert the grayscale image
    inverted_gray_image = 255 - gray_image
    # Apply Gaussian blur to the inverted grayscale image
    blurred_image = cv2.GaussianBlur(inverted_gray_image, (21, 21), 0)
    # Invert the blurred image
    inverted_blurred_image = 255 - blurred_image
    # Create the pencil sketch by dividing the grayscale image by the inverted blurred image
    pencil_sketch = cv2.divide(gray_image, inverted_blurred_image, scale=256.0)
    return pencil_sketch

# Path to the input image
image_path = '/content/drive/MyDrive/New folder/images 1.jpeg'

# Generate the pencil sketch
sketch = pencil_sketch(image_path)

from google.colab.patches import cv2_imshow

# Display the original image and the pencil sketch
cv2_imshow(cv2.imread(image_path))
cv2_imshow(sketch)


cv2.waitKey(0)
cv2.destroyAllWindows()
```
